## Linear Regression Using Least Square Method

(Step 1) We need the following module

```python
import numpy as np
from sklearn.linear_model import LinearRegression, SGDRegressor
```

(Step 2) Define the input feature and output variables

```python
# X = np.array([1, 2, 3, 4, 5])
# y = np.array([2, 4, 5, 4, 5])

X = np.array([4, 6, 7, 9, 1, 6, 14, 15, 4, 7])
y = np.array([2.45, 3.12, 2.79, 3.08, 1.99, 2.19, 4, 3.24, 3.19, 2.55])

X = X.reshape((-1,1)) # Reshape the input feature into a 2D array
y = y.reshape((-1,1)) # Reshape the input feature into a 2D array
```

(Step 3) Create and train the linear regression model

```python
model = LinearRegression()
model.fit(X,y)
w1 = model.coef_.item()
w0 = model.intercept_
print("w0 = ",w0)
print("w1 = ", w1)
```

(Step 4) Display the training data and created model

```python
import matplotlib.pyplot as plt
plt.scatter(X, y)
plt.plot(X, w1 * X + w0, 'r')
plt.xlabel("X")
plt.ylabel("y")
plt.title("Linear Regression using Least Squares Method")
plt.show()
```

(Step 5) Predict the output variable for new input feature values

```python
new_x = np.array([4]).reshape((-1, 1))
predicted_y = model.predict(new_x)

print(predicted_y)
```

(Step 2) Create a Gradient Descent model using the Stochastic Gradient Descent method

```python
model = SGDRegressor(loss='squared_error', alpha=0.01, learning_rate='constant', max_iter=2000)

# Fit the model to the data
model.fit(X, y)

print(model.coef_)
print(model.intercept_)
w0=model.intercept_
w1=model.coef_
```
