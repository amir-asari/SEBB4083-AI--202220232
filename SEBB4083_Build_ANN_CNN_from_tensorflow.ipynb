## This is a few example how to built and display the CNN or ANN model in tensorflow

### Setup

```python
import tensorflow as tf
from tensorflow.keras import layers, models
```

### Example 1
#### 2-layer NN

Layer 1: Input layer with 3 inputs

Layer 2: Hidden layer with 4 neurons/nods (activation function-sigmoid)

Layer 3: Output layer with 2 neurons/nods (activation function-sigmoid)

Screenshot 2023-05-24 225934.png

```python
model = models.Sequential(name = '2-layer NN')
model.add(layers.Dense(4, activation = 'sigmoid', input_dim = 3, name = 'hidden_layer'))
model.add(layers.Dense(2, activation = 'sigmoid', name = 'output_layer'))
```

There are 2 methods to display the architecture of the developed model:

1) using summary function

2) using plot_model function

```python
# Method 1
model.summary()
```

```python
# Method 2
tf.keras.utils.plot_model(model, to_file = '2-layer NN.png', show_shapes = True)
```

### Example 2
#### 3-layer NN

Layer 1: Input layer with 3 inputs

Layer 2: Hidden layer with 4 neurons/nods (activation function-sigmoid)

Layer 3: Hidden layer with 4 neurons/nods (activation function-sigmoid)

Layer 4: Output layer with 2 neurons/nods (activation function-sigmoid)

Screenshot 2023-05-24 233032.png

```python
model = models.Sequential(name = '3-layer NN')
model.add(layers.Dense(4, activation = 'sigmoid', input_dim = 3, name = 'hidden_layer1'))
model.add(layers.Dense(4, activation = 'sigmoid', input_dim = 3, name = 'hidden_layer2'))
model.add(layers.Dense(2, activation = 'sigmoid', name = 'output_layer'))
```

```python
# Method 1
model.summary()
```

```python
# Method 2
tf.keras.utils.plot_model(model, to_file = '3-layer NN.png', show_shapes = True)
```

### Example 3
#### Demonstrate CNN layer only

Layer 1: Input layer with 32 x 32 Grayscale Image

Layer 2: Convolution layer with 6 filters 28x28, stride = 1, no padding  (activation function-ReLU)

```python
model = models.Sequential(name = 'SimpleCNN')
model.add(layers.Conv2D(6, (3, 3), activation='relu', strides = 1, input_shape=(32, 32, 1), padding = 'valid' ,name='Conv1'))
```

```python
# Method 1
model.summary()
```

```python
# Method 2
tf.keras.utils.plot_model(model, to_file = 'SimpleCNN.png', show_shapes = True)
```

### Example 4
#### Demonstrate one CNN layer and one pooling layer

Layer 1: Input layer with 32 x 32 Grayscale Image

Layer 2: Convolution layer with 6 filters 28x28, stride = 1, no padding  (activation function-ReLU)

Layer 3: Max pool layer with 2x2 filter, stride =1, no padding 

```python
model = models.Sequential(name = 'SimpleCNN')
model.add(layers.Conv2D(6, (3, 3), activation='relu', strides = 1, input_shape=(32, 32, 1), padding = 'valid' ,name='Conv1'))
model.add(layers.MaxPooling2D((2, 2), strides = 2, name='MaxPool1'))
```

```python
# Method 1
model.summary()
```

```python
# Method 2
tf.keras.utils.plot_model(model, to_file = 'SimpleCNN.png', show_shapes = True)
```

### Example 5
#### CNN - Simple CNN

Layer 1: Input layer with 28 x 28 Grayscale Image

Layer 2: Convolution layer with 6 filters 28x28, stride = 1, padding  (activation function-ReLU)

Layer 3: Max pool layer with 2x2 filter, stride =1, no padding 

Layer 4: Fully Connected Layer with 6 neurons (activation function-ReLU)

Layer 5: Fully Connected Layer with 3 neurons 

Simple CNN.jpg

```python
model = models.Sequential(name = 'SimpleCNN')
model.add(layers.Conv2D(6, (3, 3), activation='relu', strides = 1, input_shape=(28, 28, 1), padding = 'same' ,name='Conv1'))
model.add(layers.MaxPooling2D((2, 2), strides = 2, name='MaxPool1'))
model.add(layers.Flatten())
model.add(layers.Dense(6, activation ='relu',name ='Dense1'))
model.add(layers.Dense(3, name ='Dense2'))
```

```python
# Method 1
model.summary()
```

```python
# Method 2
tf.keras.utils.plot_model(model, to_file = 'SimpleCNN.png', show_shapes = True)
```

### Example 6
#### CNN

Layer 1: Input layer with 32 x 32 Grayscale Image

Layer 2: Convolution layer with 6 filters 5x5, stride = 1, no padding  (activation function-ReLU)

Layer 3: Max pool layer with 2x2 filter, stride =1, no padding 

Layer 4: Convolution layer with 16 filters 5x5, stride = 1, no padding  (activation function-ReLU)

Layer 5: Max pool layer with 2x2 filter, stride =1, no padding 

Layer 6: Fully Connected Layer with 120 neurons (activation function-ReLU)

Layer 7: Fully Connected Layer with 84 neurons (activation function-ReLU)

Layer 8: Fully Connected Layer with 10 neurons

The-architecture-of-LeNet-5-23-a-CNN-used-for-digits-recognition-for-the-MNIST-dataset.png

```python
model = models.Sequential(name = 'Le-Net5')
model.add(layers.Conv2D(6, (5, 5), activation='relu', strides=1, input_shape=(32, 32, 1), padding = 'valid' ,name='Conv1'))
model.add(layers.MaxPooling2D((2, 2),strides=2, name='MaxPool1'))
model.add(layers.Conv2D(16, (5, 5), activation='relu', strides=1, padding = 'valid' ,name='Conv2'))
model.add(layers.MaxPooling2D((2, 2), strides=2, name ='MaxPool2'))
model.add(layers.Flatten())
model.add(layers.Dense(120, activation ='relu',name ='Dense1'))
model.add(layers.Dense(84, activation ='relu',name ='Dense2'))
model.add(layers.Dense(10, name ='Dense3'))
```

```python
# Method 1
model.summary()
```

```python
# Method 2
tf.keras.utils.plot_model(model, to_file = 'Le-Net5.png', show_shapes = True)
```
