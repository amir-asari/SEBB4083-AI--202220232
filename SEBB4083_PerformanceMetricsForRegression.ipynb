Examples

Case 1 Model gives accurate results

case1.png

R2 = 1- 0/200 = 1

```python
from sklearn.metrics import r2_score
import matplotlib.pyplot as plt
import numpy as np

### Assume y is the actual value and f is the predicted values
# y =np.array([10, 20, 30])
# f =np.array([10, 20, 30])
y =np.array([10, 20, 30])
f =np.array([15, 18, 32])

r2 = r2_score(y, f)
print('r2 score for perfect model is', r2)
```

```python
# Plot the data and regression line
plt.scatter(y, f, alpha=0.5, label='Actual Data')
plt.plot(y, y, color='red', label='Regression Line')
plt.title(f'R2 Score: {r2:.2f}')
plt.legend()
plt.show()
```

```python
# Obtaining other performance metrics, SSE, MSE, MAE and RMSE
from sklearn.metrics import mean_squared_error, mean_absolute_error


# Compute MSE
mse = mean_squared_error(y, f)
print("MSE:", mse)

# Compute MAE
mae = mean_absolute_error(y, f)
print("MAE:", mae)

# Calculate MSE using numpy
mse = np.mean((f - y)**2)

print("MSE:", mse)

# Calculate RMSE using numpy
rmse = np.sqrt(mse)

print("RMSE:", rmse)
```
