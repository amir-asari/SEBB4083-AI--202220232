##Linear Regression Using Least Square Method (using 1 dimensional feature)

(Step 1) We need 2 module/library 
1.   numpy - managing the array and math operation
2.   matplotlib - display the data

```python
import numpy as np
import matplotlib.pyplot as plt
```

(Step 2) Assume the following is our dataset

```python
# # Define the dataset
# X = np.array([1.2, 1.8, 3, 4, 5])
# y = np.array([2.1, 4, 6, 8, 10])

# # Generate dummy dataset
# np.random.seed(42)
# X = np.linspace(0, 10, num=5)
# y = 2*X + 1 + np.random.normal(0, 2, size=X.shape)

# # Define the dataset
X = np.array([4, 6, 7, 9, 1, 6, 14, 15, 4, 7])
y = np.array([2.45, 3.12, 2.79, 3.08, 1.99, 2.19, 4, 3.24, 3.19, 2.55])

# X = np.array([2, 8, 11, 5, 3, 10, 13, 12, 9, 6])
# y = np.array([1.56, 3.25, 2.87, 2.11, 1.76, 3.01, 3.68, 3.48, 2.92, 2.39])

```

(Step 3) Obtain $\bar{x}$ and $\bar{y}$, which are mean of x and y

```python
# Compute the mean of X and y
mean_x = np.mean(X)
mean_y = np.mean(y)
print('mean x=',mean_x)
print('mean y=', mean_y)
```

(Step 4) Compute
$\left( x_{i}-\bar{x} \right)$, represent by X_diff; and $\left( y_{i}-\bar{y} \right)$, represent by y_diff
in the following code

```python
# Compute the difference between each data point and the mean
X_diff = X - mean_x
y_diff = y - mean_y
print(X_diff)
print(y_diff)
```

Compute:


*   $\sum_{i=1}^{N}\left( x_{i}-\bar{x} \right)\left( y_{i}-\bar{y} \right)$, represent by SS_xy
*   $\sum_{i=1}^{N}\left( x_{i}-\bar{x} \right)^2$, represent by SS_xx\
in the following code



```python
# Compute the sum of squared differences
SS_xx = np.sum(X_diff ** 2)
SS_xy = np.sum(X_diff * y_diff)

# np.set_printoptions(suppress=True)
# np.set_printoptions(suppress=False)
# SS_xx=np.round(X_diff ** 2,3)
# print(SS_xx)
# SS_xx = np.sum(SS_xx)
# print(SS_xx)
# SS_xy=np.round(X_diff * y_diff,3)
# print(SS_xy)
# SS_xy = np.sum(SS_xy)
# print(SS_xy)

print(SS_xx)
print(SS_xy)
```

Finally,

$\hat{w}=\frac{\sum_{i=1}^{N}\left( x_{i}-\bar{x} \right)\left( y_{i}-\bar{y} \right)}{\sum_{i=1}^{N}\left( x_{i}-\bar{x} \right)^2}$

and

$\hat{b}=\bar{y}-w\bar{x}$

```python
# Compute the slope and intercept using least squares method
w = SS_xy / SS_xx #slope
b = mean_y - w * mean_x #intercept

# Print the coefficients
print("Slope, w: {:.2f}".format(w))
print("Intercept, b: {:.2f}".format(b))
print("Model produced: y=%.2fx+%.2f" %(w, b))

# Plot the data and the regression line
plt.scatter(X, y)
plt.plot(X, w * X + b, 'r')
plt.xlabel("X")
plt.ylabel("y")
plt.title("Linear Regression using Least Squares Method")
plt.show()
```

##Linear Regression Using Gradient Descent (using 1 dimensional feature)

```python
import numpy as np
import matplotlib.pyplot as plt

# # Generate dummy dataset
# np.random.seed(42)
# X = np.linspace(0, 10, num=50)
# y = 2*X + 1 + np.random.normal(0, 2, size=X.shape)

# Define the dataset
X = np.array([4, 6, 7, 9, 1, 6, 14, 15, 4, 7])
y = np.array([2.45, 3.12, 2.79, 3.08, 1.99, 2.19, 4, 3.24, 3.19, 2.55])

# Define the learning rate and number of iterations
learning_rate = 0.01
n_iterations = 1000

# Initialize the slope and intercept to 0
w1 = 0
w0 = 0

# Define the number of samples
n = float(len(X))

# Perform gradient descent
for i in range(n_iterations):
    y_pred = X*w1 + w0
    w0_gradient = (-1/n) * sum(y - y_pred)
    w1_gradient = (-1/n) * sum(X * (y - y_pred))
    w0 -= learning_rate * w0_gradient
    w1 -= learning_rate * w1_gradient
    

# Print the coefficients
print("Slope w1: {:.2f}".format(w1))
print("Intercept w0: {:.2f}".format(w0))
print("Model produced: y=%.2fx+%.2f" %(w1, w0))

# Plot the data and the regression line
plt.scatter(X, y)
plt.plot(X, w1 * X + w0, 'r')
plt.xlabel("X")
plt.ylabel("y")
plt.title("Linear Regression using Gradient Descent")
plt.show()
```

##Linear Regression Using Gradient Descent (using 1 dimensional feature) - in vector form

```python
import numpy as np
import matplotlib.pyplot as plt


# # Define the dataset
X = np.array([4, 6, 7, 9, 1, 6, 14, 15, 4, 7])
y = np.array([[2.45, 3.12, 2.79, 3.08, 1.99, 2.19, 4, 3.24, 3.19, 2.55]])

X=np.transpose(X).astype(float);
y=np.transpose(y);
# print(y)


# Define the learning rate and number of iterations
learning_rate = 0.01
n_iterations = 1000

#Define the number of feature dimension
D=1

# Initialize the parameters/weights in vector w to zero (w0-intercept, w1-slope)
w = np.zeros((D+1,1))

# Define the number of samples
N = float(len(X))

#add a column of 1's for the bias term
X_bias=np.c_[np.ones((int(N), 1)), X]
# print(X_bias)


# Perform gradient descent
for i in range(n_iterations):
  # e1 = np.dot(X_bias, w) 
  gradients = (1/N) * np.dot(np.transpose(X_bias), (np.dot(X_bias, w) - y))
  # print(gradients)
  w -= learning_rate * gradients

    

# Print the coefficients
for i in range(D+1):
  print("Parameter w%d = %.2f" %(i, w[i].item()))


# # Plot the data and the regression line
plt.scatter(X, y)
plt.plot(X, w[1].item() * X + w[0].item(), 'r')
plt.xlabel("X")
plt.ylabel("y")
plt.title("Linear Regression using Gradient Descent")
plt.show()
```
